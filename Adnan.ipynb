```python
class CSE_FEST:
    def __init__(self):
        global cnt
        cnt += 1

    def start(self):
        print("CSE FEST".center(68))
        print("_" * 120)
        print()


class First_day(CSE_FEST):
    def first_day(self):
        print("Let's talk about first day")
        print()


class Gaming(First_day):
    def __init__(self):
        print("Gaming Section: ")

    @staticmethod
    def fifa_points_table(name, points):
        with open("Fest.txt", "a") as fest:
            fest.write("Fifa contestants' name and their scores are Given below: \n\n")
            fest.write("{:<10} {:<10}\n".format("Name", "Points"))
            for n, p in zip(name, points):
                fest.write("{:<10} {:<10}\n".format(n, p))

    @staticmethod
    def fifa_winner(name, points):
        points_dict = dict(zip(name, points))
        winner = max(points_dict, key=points_dict.get)
        with open("Fest.txt", "a") as fest:
            fest.write("\nWinner is: {}\n".format(winner))

    @staticmethod
    def c_make_fixture(name):
        print("\nLet's make an interesting thing. Let's make a fixture of the chess tournament :3\n")
        for i in range(4):
            for j in range(4):
                if i != j:
                    print(name[i], "VS", name[j], "\t")
            print()


def valorant_points_table(name, points):
    with open("Fest.txt", "a") as fest:
        fest.write("\nValorant Teams and their scores are Given below: \n\n")
        fest.write("{:<10} {:<10}\n".format("Team Name", "Points"))
        for n, p in zip(name, points):
            fest.write("{:<8} {:<11}\n".format(n, p))


def valorant_winner(name, points):
    points_dict = dict(zip(name, points))
    winner = max(points_dict, key=points_dict.get)
    with open("Fest.txt", "a") as fest:
        fest.write("\nWinner is: {}\n".format(winner))


def chess_points_table(name, points):
    with open("Fest.txt", "a") as fest:
        fest.write("\nChess players and their scores are Given below: \n\n")
        fest.write("{:<10} {:<10}\n".format("Name", "Points"))
        for n, p in zip(name, points):
            fest.write("{:<10} {:<10}\n".format(n, p))


def chess_winner(name, points):
    points_dict = dict(zip(name, points))
    winner = max(points_dict, key=points_dict.get)
    with open("Fest.txt", "a") as fest:
        fest.write("\nWinner is: {}\n".format(winner))


class Second_day(CSE_FEST):
    def second_day(self):
        print("Let's talk about second day")
        print()


class Coding(Second_day):
    def __init__(self):
        print("Coding Section: \n")

    @staticmethod
    def round_winner(name, time):
        points_dict = dict(zip(name, time))
        winner = min(points_dict, key=points_dict.get)
        return winner

    @staticmethod
    def duel_points_table(name, points):
        with open("Fest.txt", "a") as fest:
            fest.write("{:<10} {:<10}\n".format("Name", "Points"))
            for n, p in zip(name, points):
                fest.write("{:<9} {:<9}\n".format(n, p))

    @staticmethod
    def ranking(name, points, time):
        vpn = [(p, n) for n, p in zip(name, points)]
        vtn = [(t, n) for n, t in zip(name, time)]
        vpn.sort(reverse=True)
        vtn.sort()
        with open("Fest.txt", "a") as fest:
            fest.write("{:<10} {:<23} {:<7}\n".format("TEAM NAME", "No of solved Problem", "TIME"))
            for i in range(10):
                fest.write("{:<8} {:<14} {:<17}\n".format(vpn[i][1], vpn[i][0], vtn[i][0]))
            fest.write("\nWinner is {}\n".format(vpn[0][1]))


if __name__ == "__main__":
    with open("Fest.txt", "w") as fest:
        fest.write("Hi CSE_FEST\n\n")

    cnt = 0
    start = CSE_FEST()
    FD = First_day()

    start.start()
    FD.first_day()

    a = Gaming()

    print("\nFIFA: ")
    F_name = []
    F_points = []
    for _ in range(4):
        s, x = input().split()
        F_name.append(s)
        F_points.append(int(x))

    a.fifa_points_table(F_name, F_points)
    a.fifa_winner(F_name, F_points)

    with open("Fest.txt", "a") as fest:
        fest.write("\n")

    start.start()
    print("Valorant: ")
    V_name = []
    V_points = []
    for _ in range(4):
        s, x = input().split()
        V_name.append(s)
        V_points.append(int(x))

    valorant_points_table(V_name, V_points)
    valorant_winner(V_name, V_points)

    with open("Fest.txt", "a") as fest:
        fest.write("\n")

    start.start()
    print("Chess: ")
    C_name = []
    C_points = []
    for _ in range(4):
        s = input()
        C_name.append(s)

    a.c_make_fixture(C_name)
    print("\nPoints accordingly")
    for _ in range(4):
        s = int(input())
        C_points.append(s)

    chess_points_table(C_name, C_points)
    chess_winner(C_name, C_points)

    with open("Fest.txt", "a") as fest:
        fest.write("\n")

    start.start()
    SD = Second_day()
    SD.second_day()

    b = Coding()

    print("\nDUEL Compettition: ")
    D_names = [[] for _ in range(4)]
    D_Final = []
    t_Final = []

    for i in range(4):
        print("Batch", i + 17, ": \n")
        for j in range(4):
            name, time = input().split()
            D_names[i].append(name)
            if i == 0:
                D_Final.append([])
            D_Final[i].append(name)
            t_Final.append(int(time))
        print()

    with open("Fest.txt", "a") as fest:
        fest.write("\nDuel competition and time table\n")
        for i in range(4):
            fest.write("Batch {}: \n".format(i + 17))
            b.duel_points_table(D_names[i], t_Final[i])

    print("final round: \nNeeded times for each winner given in batch descending order")
    with open("Fest.txt", "a") as fest:
        fest.write("\nFinal Table: \n")
    b.duel_points_table(D_Final, t_Final)

    with open("Fest.txt", "a") as fest:
        fest.write("\nFinally the Champion is {}\n\n".format(b.round_winner(D_Final, t_Final)))

    start.start()
    print("Programming Contest\n")
    with open("Fest.txt", "a") as fest:
        fest.write("Programming Contest\n\n")
        fest.write("RANKING: \n")

    P_Team = []
    No_of_solved_problem = []
    P_time = []
    for _ in range(10):
        team, solved, time = input().split()
        P_Team.append(team)
        No_of_solved_problem.append(int(solved))
        P_time.append(int(time))

    b.ranking(P_Team, No_of_solved_problem, P_time)

    print("Here {} events have been organized by Dept. Of CSE, CUET. If you want to see the summary, then please go to fest.txt file".format(cnt))
```
